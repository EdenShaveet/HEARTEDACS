```python
#PREREQUISITES
alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'] #create alphabet character list for input quality control
```

```python
#USER INPUTS
Flag=True #set while loop flag
while Flag==True:
  age=input("Patient Age (in years):\n")
  for i in age:
    if i in alphabet:
      print("Error. Please enter the patient's age as a whole number between 18 and 200.")
      break #terminate for loop
    else:
      age=int(age)
      if age>200 or age<18:
        print("Error. Please enter the patient's age as a whole number between 18 and 200.")
        break
      else:
        Flag=False #change flag to terminate while loop
Flag=True
while Flag==True:
  sex=input("Patient Sex (male or female):\n").lower()
  if sex!="male" and sex!="female":
    print("Error. Please enter 'male' or 'female'")
  else:
    Flag=False
Flag=True
while Flag==True:
  history=input("Patient History Score (0-2):\n")
  for i in history:
    if i in alphabet:
      print("Error. Please enter a score between 0 and 2.")
      break
    else:
      history=int(history)
      if history!=0 and history!=1 and history!=2:
        print("Error. Please enter a score between 0 and 2.")
        break
      else:
        Flag=False
Flag=True
while Flag==True:
  ecg=input("ECG Score (0-2):\n")
  for i in ecg:
    if i in alphabet:
      print("Error. Please enter a score between 0 and 2.")
      break
    else:
      ecg=int(ecg)
      if ecg!=0 and ecg!=1 and ecg!=2:
        print("Error. Please enter a score between 0 and 2.")
        break
      else:
        Flag=False
Flag=True
while Flag==True:
  risk=input("Risk Factor Score (0-2):\n")
  for i in risk:
    if i in alphabet:
      print("Error. Please enter a score between 0 and 2.")
      break
    else:
      risk=int(risk)
      if risk!=0 and risk!=1 and risk!=2:
        print("Error. Please enter a score between 0 and 2.")
        break
      else:
        Flag=False
Flag=True
while Flag==True:
  trop=input("Troponin Score (0-2):\n")
  for i in trop:
    if i in alphabet:
      print("Error. Please enter a score between 0 and 2.")
      break
    else:
      trop=int(trop)
      if trop!=0 and trop!=1 and trop!=2:
        print("Error. Please enter a score between 0 and 2.")
        break
      else:
        Flag=False
Flag=True
while Flag==True:
  dia=input("Is the patient experiencing diaphoresis? (yes or no)\n").lower()
  if dia!="yes" and dia!="no":
    print("Error. Please enter 'yes' or 'no'")
  else:
    Flag=False
Flag=True
while Flag==True:
  pain=input("Is the patient experiencing pain that radiates to their arm or shoulder? (yes or no)\n").lower()
  if pain!="yes" and pain!="no":
    print("Error. Please enter 'yes' or 'no'")
  else:
    Flag=False
Flag=True
while Flag==True:
  worse=input("Does the patient's pain occur or worsen with inspiration? (yes or no)\n").lower()
  if worse!="yes" and worse!="no":
    print("Error. Please enter 'yes' or 'no'")
  else:
    Flag=False
Flag=True
while Flag==True:
  palp=input("Is the patient's pain reproduced by palpitation? (yes or no)\n").lower()
  if palp!="yes" and palp!="no":
    print("Error. Please enter 'yes' or 'no'")
  else:
    Flag=False
```

```python
#EDACS CALCULATION
edacs=0 #set score to zero

#Age
if age<46:
  edacs+=2
elif age<51:
  edacs+=4
elif age<56:
  edacs+=6
elif age<61:
  edacs+=8
elif age<66:
  edacs+=10
elif age<71:
  edacs+=12
elif age<76:
  edacs+=14
elif age<81:
  edacs+=16
elif age<86:
  edacs+=18
else:
  edacs+=20

#Sex
if sex=="male":
  edacs+=6
else:
  edacs+=0

#Risks
if age<51 and risk==2:
  edacs+=4
else:
  edacs+=0

#Y/N questions
if dia=="yes":
  edacs+=3
if pain=="yes":
  edacs+=5
if worse=="yes":
  edacs-=4
if palp=="yes":
  edacs-=6
else:
  edacs+=0
```

```python
#EDACS OUTPUT
if edacs<16 and ecg==0 and trop==0:
  print(f"The patient's EDACS is {edacs} (low risk).")
elif edacs<16 and ecg!=0 or edacs<16 and trop!=0:
  print(f"The patient's EDACS is {edacs} (low risk); however additional information provided about ECG and troponin may necessitate further observation.")
elif edacs>15:
  print(f"The patient's EDACS is {edacs} (NOT low risk).")
else:
  print("Something went wrong. Try again.")
```

```python
#HEART SCORE CALCULATION
heart=0 #set score to zero

#Age
if age<46:
  heart+=0
elif age<66:
  heart+=1
else:
  heart+=2

#Integer input addition
heart+=history
heart+=ecg
heart+=risk
heart+=trop
```

```python
#HEART SCORE OUTPUT
if heart<4:
  print(f"The patient's HEART Score is {heart}. They are at LOW risk (1.6%) of reaching a major adverse cardiac event (MACE) in the next 30 days.")
elif heart<7:
  print(f"The patient's HEART Score is {heart}. They are at MEDIUM risk (13%) of reaching a major adverse cardiac event (MACE) in the next 30 days.")
elif heart>6:
  print(f"The patient's HEART Score is {heart}. They are at HIGH risk (50%) of reaching a major adverse cardiac event (MACE) in the next 30 days.")
else:
  print("Something went wrong. Try again.")
```

```python
#JOINT OUTPUT
if edacs<16 and heart<4:
  print(f"Both the patient's HEART score ({heart}) and EDACS ({edacs}) indicate low risk.")
elif edacs>15 and heart>3:
  print(f"Both the patient's HEART score ({heart}) and EDACS ({edacs}) indicate non-low risk.")
else:
  print("Refer to individual guidence provided by EDACS and HEART Score documentation.")
```
